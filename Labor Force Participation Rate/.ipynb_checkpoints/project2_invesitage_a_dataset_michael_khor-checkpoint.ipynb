Abbreviations:

_Here are some questions I decide to ask regarding their Work > Labor Force Participation Rate data:_
1. What is the distribution of workers by age for top and bottom countries in GDP per capita? (bar graph, time-based)
2. Through the years, which region of the world has the highest age 65+ LFPR; highest women LFPR?  (line graphs)
3. Is there a correlation between women LFPR, age 65+ LFPR, overall LFPR, and country GDP and GDP growth? (scatter plots, matrix)

A list of data tables used is available at the end of notebook.

## Data Import

```python
# Import relevant packages
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

# Set up categories of LFPR data
ages_csv = ['15plus', '15_64', '25_54', '65plus']
genders_csv = ['', 'females_', 'males_']
genders_label = ['all', 'female', 'male']

# Open all LFPR csv's using a loop
workdata = dict()
for gender in genders_csv:
    for age in ages_csv:
        workdata[genders_label[genders_csv.index(gender)]+'_'+age] = pd.read_csv('datasets/'+gender + 'aged_' + age + '_labour_force_participation_rate_percent.csv')

# Open all other csv's individually
gdp_data = dict()
gdp_data['gdp_per_cap'] = pd.read_csv('datasets/gdppercapita_us_inflation_adjusted.csv')
gdp_data['gdp_growth'] = pd.read_csv('datasets/gdp_per_capita_yearly_growth.csv')
data_geography = pd.read_csv('datasets/Data Geographies - v1 - by Gapminder - list-of-countries-etc.csv')

# Preview of LFPR dataframe
print("Here's a preview of how each LFPR dataframe looks like:")
workdata['all_15plus'].head(1)
```

## Data Wrangling

First, I melt all workers data (LFPR) and combine them into one dataframe. I also associate a world region to each country.

```python
# Melt dataframe from multiple year columns into just 3 columns, then merge into single table
workdata_melt = dict()
workdata_concat = pd.DataFrame(columns=['country','year','lfpr','age', 'gender'])
for gender in genders_label:
    for age in ages_csv:
        work_years = workdata[gender+'_'+age].columns.drop('country')
        workdata_melt[gender+'_'+age] = workdata[gender+'_'+age].melt(id_vars='country', var_name='year', value_name='lfpr')
        workdata_melt[gender+'_'+age]['year'] = workdata_melt[gender+'_'+age]['year'].astype({'year':'int'})
        workdata_melt[gender+'_'+age]['age'] = age
        workdata_melt[gender+'_'+age]['gender'] = gender
        workdata_concat = pd.concat([workdata_concat, workdata_melt[gender+'_'+age] ]);

workdata_concat['year'] = workdata_concat['year'].astype(int)
print('Each csv contained',workdata_melt[gender+'_'+age].shape[0],'elements.')
print('After concatenation, there should be',workdata_melt[gender+'_'+age].shape[0]*len(workdata.keys()), 'rows.')

# Adding feature world region to dataframe
world_regions = data_geography.sort_values('eight_regions')['eight_regions'].unique().tolist()
workdata_concat['region'] = workdata_concat['country'].map(data_geography.set_index('name')['eight_regions'])
print('There are',workdata_concat.shape[0],'rows in workdata_concat.')
workdata_concat.head(1)
```

Next, I make a subset of the master dataframe for only the top and bottom GDP countries (in 2017) --> for Question 1

```python
# Clean GDP data to ensure that it has workers data
gdp_data['gdp_per_cap'] = gdp_data['gdp_per_cap'][gdp_data['gdp_per_cap']['country'].isin(workdata_concat['country'].unique())]

# Query for top and bottom GDP countries, make a list
n_countries = 30
order_ascending_gdp = [False, True]
order_gdp = ['top','bottom']
country_list = dict()
country_workers = dict()
for n in range(len(order_gdp)):
    country_list[order_gdp[n]] = gdp_data['gdp_per_cap'].sort_values('2017', ascending=order_ascending_gdp[n]).head(n_countries)['country'].to_list()
    country_workers[order_gdp[n]] = workdata_concat.query("country in {} & gender == 'all' & year == 2017".format(country_list[order_gdp[n]]))
    country_workers[order_gdp[n]].loc[:,'gdp_category'] = order_gdp[n]

country_list['all'] = country_list['top']+country_list['bottom']

# Concatenating both dataframes into one, then sort by age group name
country_workers = pd.concat([country_workers['top'],country_workers['bottom']])
country_workers.sort_values('age', inplace=True);
```

Here, I create a dataframe with mean LFPRs based on world region, age group, gender, and year --> for Question 2

```python
# Check for empty data
print('Number of empty rows:',sum(workdata_concat['lfpr'].isnull()))

# Create a dataframe with mean LFPR for each combination of groups
workdata_mean = workdata_concat.groupby(['region','age','gender','year'], as_index=False).mean()
workdata_mean.head(1)
```

Lastly, I create a dataframe on workers data combined with GDP data to get correlations --> for Question 3

```python
# Make a copy of workdata to be merged with GDP data
workdata_gdp = workdata_concat.copy()

gdp_features = ['gdp_per_cap','gdp_growth']
for feature in gdp_features:
    # Melt GDP dataframe into a table with 3 columns without empty data
    gdp_melt = gdp_data[feature].melt(id_vars = 'country', value_name=feature, var_name='year').dropna()
    gdp_melt['year'] = gdp_melt['year'].astype(int)

    # Set multiindex for matching
    gdp_melt.set_index(['country', 'year'], inplace=True)
    workdata_gdp.set_index(['country','year'], inplace=True)

    # Create column for gdp per capita in workdata_gdp dataframe, drop empty/non-matching rows
    workdata_gdp[feature] = workdata_gdp.index
    workdata_gdp[feature] = workdata_gdp[feature].map(gdp_melt[feature])
    workdata_gdp.dropna(inplace=True)
    workdata_gdp.reset_index(inplace=True)
    
workdata_gdp.head(2)
```

## Data Visualization

### Question 1: distribution of workers by age for top and bottom countries in GDP

I'm curious to see if certain age groups' labor force participation rate (particularly with older working population) differ in richer vs poorer countries. For that, I will be only using top 30 and bottom 30 countries that I queried out, and compare LFPRs between those two groups of countries by age.

```python
# Create bar chart for each age group
position = np.arange(len(ages_csv))
ticklabels = country_workers['age'].str.replace('_',' to ').unique().tolist()
width = (1 - 0.2)/len(country_list['all'])
figure_all = plt.figure(figsize = [13,6])

for i in range(0,len(country_list['all'])):
    heights = country_workers.query("country == '{}'".format(country_list['all'][i]))['lfpr'].to_list()
    color = 'r' if (country_list['all'][i] in country_list['top']) else 'b'
    figure_all = plt.bar(position+i*width, heights, width = width, color=color);

plt.title('Work participation of top GDP/capita countries (red) and bottom GDP/capita countries (blue) by age group')
plt.ylabel('Labor Force Participation Rate')
plt.xlim(-0.2,5)
#plt.legend(country_list, loc='upper right')
plt.xticks(ticks=position+0.4, labels=ticklabels);
```

Based on the graph above, I can draw two observations:
1. For the age group 15-to-64, countries with low GDP (in blue) generally have a lower LFPR than high GDP countries (red).
2. For the age group of 65 and above, the opposite is true, where low GDP countries have a significantly higher LFPR than high GDP countries.

However, the observation made was "eye-balled". To get a more definite analysis, I take the mean of each GDP countries group, and plot the LFPR by age group.

```python
country_work_mean = country_workers.groupby(['gdp_category','age'],as_index=False).mean()

# Create bar chart for each age group
position = np.arange(len(ages_csv)) + 0.5
ticklabels = country_work_mean['age'].str.replace('_',' to ').unique().tolist()
width = (1 - 0.2)/2
figure_mean = plt.figure(figsize = [13,6])

for i in range(0,2):
    heights = country_work_mean.query("gdp_category == '{}'".format(order_gdp[i]))['lfpr'].to_list()
    color = 'r' if (order_gdp[i] == 'top') else 'b'
    figure_mean = plt.bar(position+i*width, heights, width = width, color=color);

plt.title('Mean work participation of top GDP/capita countries (red) and bottom GDP/capita countries (blue) by age group')
plt.ylabel('Labor Force Participation Rate')
plt.xlim(-0.2,5)
plt.legend(order_gdp, loc='upper right')
plt.xticks(ticks=position+0.2, labels=ticklabels);
```

Now we can definitively support the two observations mentioned above.

I must acknowledge that initially my pre-analysis assumption is that higher LFPR from the aging population will contribute to the overall economy of a country, hence lead to a higher GDP per capita. However, another explanation is more likely-- where countries with lower GDP per capita forces its people to retire later than other wealthier countries, hence inflating the LFPR in the age category of 65+.

### Question 2: Which region of the world has the highest age 65+ LFPR; highest women LFPR?

I wonder how labor force participation rate for specific demographics (older population; women) has trended over the years for each world region. For that, I decided to plot a year-based line graph of each desired LFPR, grouped and averaged by world regions.

```python
# Plot line graph for various criteria
subset_queries = ["age == '65plus' & gender == 'all'",
                 "age == '15plus' & gender == 'female'"]
subset_titles = ['Labor Force Participation for ages 65+ by world regions',
                'Labor Force Participation for female population by world regions']
for p in range(len(subset_queries)):
    workdata_subset = workdata_mean.query(subset_queries[p])
    plt.figure(figsize=[12,6])
    for region in world_regions:
        sns.lineplot(x = workdata_subset.query("region == '{}'".format(region))['year'],
                    y = workdata_subset.query("region == '{}'".format(region))['lfpr'],
                    legend='full');
        plt.vlines(2019, workdata_subset['lfpr'].min(), workdata_subset['lfpr'].max(), linestyles='dashed')
        plt.title(subset_titles[p])
        plt.xlabel('Year')
        plt.ylabel('Labor Force Participation Rate')
        plt.xlim(1989,2045)
        plt.legend(world_regions+['2019']);
```

I've added the 2019 line to differentiate recorded data versus predictions.

Observations of graph for __ages 65+ LFPR__:
1. Sub-Saharan Africa has the highest LFPR through time, though it has steadily decreased over time.
2. Sub-Saharan Africa, West Asia, North Africa, and East Europe has seen significant decrease in LFPR over time.
3. From 1990 to present, North Africa, East Europe, and West Europe have the lowest aging population LFPR.

Observations of graph for __female population LFPR__:
1. North America, South America, and West Europe has seen a dramatic increase in LFPR since 1990, and each has passed East Europe in women LFPR around the year 2000.
2. Since 1990 to present, North Africa has the lowest female LFPR, whereas Sub-Saharan Africa has the highest female LFPR.
3. East Europe is the only world region with a decreasing trend of female LFPR since 1990.

It's worth noting that for simplicity, the numbers shown here are not weighted by each country's population, that is, a large country like China (population 1.4 billion) has the same averaging weight as Brunei (population 429k). Hence, the LFPR shown are inaccurate. However, it does give an idea of global trends of LFPR.

### Question 3: Is there a correlation between women LFPR, age 65+ LFPR, overall LFPR, and country GDP and GDP growth?

Lastly, I would like to know if there's some correlation between various LFPRs and a country's GDP. For this analysis, I'm assuming that any single country's change in LFPR over time is also correlated to a change in the country's GDP over time. In other words, using an inflation-adjusted GDP, I am able to plot every year's LFPR and GDP on the same plot and still observe any meaningful correlation.

I used queries to filter out workdata_gdp according to which LFPR I'm interested in, then plotted GDP per capita against LFPR to observe their relationship.

Note that the data used only spans between 1990 and 2017.

```python
subset_queries = ["gender == 'female'",
                 "age == '25_54'",
                 "age == '65plus'"]
subset_titles = ["women",
                "people aged 25 to 54",
                "people aged 65 and up"]
subset_hues = ["age",
              "gender",
              "gender"]

for p in range(len(subset_queries)):
    workdata_subset = workdata_gdp.query(subset_queries[p])
    plt.figure(figsize=[12,6])
    sns.scatterplot(data=workdata_subset, x="lfpr", y="gdp_per_cap", hue=subset_hues[p], alpha=0.2);
    plt.title("Labor force participation among "+subset_titles[p]+" and its relation to GDP per capita (adjusted by inflation)")
    plt.xlabel('Labor Force Participation Rate among '+subset_titles[p])
    plt.ylabel('GDP per capita (US$)')
```

I realized quickly that just applying a single feature filter does not take into account of interactions for the other feature. For example, a query for only gender == female takes in all age groups within the female population. As a result, we might end up with a confusing overlapping plot with no obvious trends (such as the first plot).

An alternative to finding correlations is plotting for every possible combination of (gender * age). Below, I used seaborn's FacetGrid function to achieve that. I'm using "year" for color hue to observe if there are any time-based shift of GDP/LFPR. I additionally plotted the same grid plot for GDP growth (in percentage).


```python
feature_ylim = [[0,100000],[-10,10]]
grids = dict()

for i in range(len(gdp_features)):
    grids[gdp_features[i]] = sns.FacetGrid(workdata_gdp, 
                                           row='gender', 
                                           col = 'age',
                                           margin_titles=True, 
                                           hue='year',
                                           xlim=[0,100],
                                           ylim=feature_ylim[i]);
    grids[gdp_features[i]].map(sns.regplot,
                               "lfpr",
                               gdp_features[i], 
                               #color = "green",
                               fit_reg=False, 
                               scatter_kws={'alpha': 0.3, 's':5}).add_legend();
    
```

Some notable observations:
1. For ages below 65 (groups "15-64" and "25-54") across all gender categories, there are positive correlations between LFPRs and a country's GDP-per-capita.
2. For ages 65 and above across all gender categories, there are negative correlations between LFPRs and a country's GDP-per-capita. Meaning, generally a high GDP country has fewer old age workers.
3. When the aforementioned two groups are combined (age group "15 plus"), the graph shapes more like a bell, with some optimum LFPR in the center yielding the highest GDP.
4. There's no observable interactions between GDP growth and LFPRs.

A major caveat to this analysis: correlation is not causation. While it's easy to assume that higher LFPR in general leads to higher GDP, there are many more factors that can affect a country's GDP and labor force participation, such as war, natural disasters, natural resources, culture etc that are not reflected in this analysis.
